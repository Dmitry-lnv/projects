# Выявление закономерностей успешности игры

На основе исторических данных о продажах игр, оценки пользователей и экспертов, жанры и платформы необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

        
исходный dataframe содержит следующую информацию:  

| | |
| --- | --- |
|Name| название игры|
|Platform| платформа|
|Year_of_Release|год выпуска|
|Genre|жанр|
|NA_sales|продажи в Северной Америке (млн проданных копий)|
|EU_salee|продажи в ЕС|
|JP_sales|продажи в Японии|
|Other_sales|продажи в других регионах|
|Critic_Score|оценка критиков (максимум 100)|
|User_Score|оценка пользователей (максимум 10)|
|Rating|возрастной рейтинг от организации ESRB|

Расшифровка рейтинга ESRB:  
        
| | |
| --- | --- |
| E | для всех возрастных категорий, раннее название K-A|
| EC | для детей от 3 лет|
| E 10+ |	для лиц старше 10 лет|
| T |	для лиц старше 13 лет|
| M |	для лиц старше 17 лет| 
| AO | для лиц старше 18 лет|
| RP | rатегория ещё не присвоена |

```python
#дополнил:
# посмотрим как ведут себя продажи для платформ относительно года их запуска
# + для удобства сравнения отнормируем их продажи за весь период существования на 1.

platform_data =[]
plt.figure(figsize=(12,3))

data = [0] * 20


for i in pt_platform.head(11).index:
    if (i != 'wii') & (i != '3ds') & (i != 'ps4') & (i != 'pc'):
        pt_mir = pt_platform_years.loc[pt_platform_years['platform'] == i]
        pt_mir = pt_mir.reset_index(drop = True)
        
        platform_data.append([i,
                              pt_mir['year_of_release'].quantile(0.05),
                              pt_mir['year_of_release'].quantile(0.95)])
        
        plt.plot(pt_mir['year_of_release'] - pt_mir['year_of_release'].quantile(0),
                 pt_mir['sum'] / pt_mir['sum'].sum(),
                 color = 'gray')
        
        for j in range(0, len(pt_mir)):
            data[j] = data[j] + pt_mir['sum'][j]

sales_reg_start = pd.DataFrame(columns = ['total_sales'], data = data)
sales_reg_start['total_sales'] = sales_reg_start['total_sales'] / sales_reg_start['total_sales'].sum()

#display (sales_reg_start)

plt.title(f'годовой объём продаж в разбивке по платвормам')
plt.ylim(0, 0.3)
plt.legend(pt_platform.head(11).index)
plt.grid(True)

plt.plot(sales_reg_start.index,
         sales_reg_start,
         color = 'red',
         linewidth = 3)

plt.show()
```

#### Возьмите данные за соответствующий актуальный период. Актуальный период определите самостоятельно в результате исследования предыдущих вопросов. Основной фактор — эти данные помогут построить прогноз на 2017 год.

```python
pt_platform_sales = df.pivot_table(index = ['platform', 'year_of_release'],
                                   aggfunc = {'total_sales' : 'sum'})
pt_platform_sales.reset_index(inplace=True)

sales = pd.DataFrame(columns = ['platform', 'first_year', 'sales_2014', 'sales_2015', 'sales_2016'])

for i in pt_platform_sales['platform'].unique():
    first_year = pt_platform_sales.loc[pt_platform_sales['platform'] == i]['year_of_release'].min()
    
    year_2014 = pt_platform_sales.loc[(pt_platform_sales['platform'] == i) &
                                      (pt_platform_sales['year_of_release'] == 2014)]['total_sales'].sum()
    year_2015 = pt_platform_sales.loc[(pt_platform_sales['platform'] == i) &
                                      (pt_platform_sales['year_of_release'] == 2015)]['total_sales'].sum()
    year_2016 = pt_platform_sales.loc[(pt_platform_sales['platform'] == i) &
                                      (pt_platform_sales['year_of_release'] == 2016)]['total_sales'].sum()
    
    sales.loc[len(sales.index )] = [i,first_year, year_2014, year_2015, year_2016]
    

sales = sales.sort_values(by = 'sales_2016', ascending = False)
sales = sales.reset_index(drop = True)
sales = sales.loc[sales['sales_2016'] >0]
display (sales)
```

```python
current_year = 2015

for forcast_year in [2016,2017]:

    sales['c_year'] = current_year - sales['first_year']
    sales['f_year'] = forcast_year - sales['first_year']

    k = []

    for i in sales.index:
        c_num_year = sales['c_year'][i]
        f_num_year = sales['f_year'][i]

        if (c_num_year > 19) | (f_num_year > 19):
            k.append(0)
        else:
            k.append(sales_reg_start['total_sales'][f_num_year] /
                                sales_reg_start['total_sales'][c_num_year])

    sales[f'k_{forcast_year}'] = k
    sales[f'f_{forcast_year}'] = sales[f'k_{forcast_year}'] * sales['sales_2015']
    
    
sales['f_2016'][3] = (sales['sales_2014'][3] + sales['sales_2015'][3]) / 2
sales['f_2017'][3] = sales['f_2016'][3]

display(sales)
```

```python
def top5 (df, name):
    try:
        sales = (df[name].sort_values(ascending = False)
                         .reset_index(drop = False)
                         .head(5))
        
        sales.loc[ len(sales.index )] = ['other', df[name].sum() - sales[name].sum()]
        return (sales)
    except:
        return(np.nan)
```

### Шаг 4 Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

    h0: средние пользовательские рейтинги платформ Xbox One и PC равны
    h1: средние пользовательские рейтинги платформ Xbox One и PC отличаются

поскольку периоды существования платформ сильно отличаются (Xbox One вышла в ноябре 2013), то необходимо сравнивать данные за одинаковый период.

Поэтому, используем метод проверки равенства для двух независимых выборок, двусторонняя гипотеза.

```python
min_year = df.loc[df['platform'] == 'xone']['year_of_release'].min()

df_pc = df.loc[(df['year_of_release'] >= min_year) &
               (df['platform'] == 'pc') &
               (df['user_score'].isna() == False)][['name', 'user_score']]
df_pc.columns = ['name', 'user_score_pc']

df_xone = df.loc[(df['year_of_release'] >= min_year) &
                 (df['platform'] == 'xone') &
                 (df['user_score'].isna() == False)][['name', 'user_score']]
df_xone.columns = ['name', 'user_score_xone']

#display (df_pc_xone)
```

```python
array1 = df_pc['user_score_pc']
array2 = df_xone['user_score_xone']

alpha = 0.01
#result = st.ttest_rel(array1, array2, alternative = 'two-sided')
result = st.ttest_ind(array1, array2, alternative = 'two-sided', equal_var = False)


print('средний пользовательский рейтинг платформы Xbox One:', round(array1.mean(),2))
print('средний пользовательский рейтинг платформы PC:', round(array2.mean(),2))
print('p-value =', result.pvalue, '\n')

if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")

bins = np.linspace(0.1, 10, 11)
plt.figure(figsize=(12,3))
plt.hist(array1, bins, alpha = 0.5, color = 'b')
plt.hist(array2, bins, alpha = 0.5, color = 'g')

plt.plot([array1.mean(), array1.mean()], [0, 30], color = 'b')
plt.plot([array2.mean(), array2.mean()], [0, 30], color = 'lime')

plt.title('Распределение пользовательских оценок')
plt.legend(['mean Xbox One', 'mean PC','dist Xbox One','dist PC'])
plt.xlabel('user score')
plt.ylabel('count')
plt.grid(True)

plt.show()
```

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

    h0: средние пользовательские рейтинги жанров Action и Sports равны
    h1: средние пользовательские рейтинги жанров Action и Sports отличаются

используем метод проверки равенства средних для двух независимых выборок, двусторонняя гипотеза.

```python
array1 = df.loc[(df['genre'] == 'action') &
                (df['user_score'].isna() == False)]['user_score']

array2 = df.loc[(df['genre'] == 'sports') &
                (df['user_score'].isna() == False)]['user_score']

alpha = 0.01
result = st.ttest_ind(array1, array2, alternative = 'two-sided', equal_var = False)



print('средний пользовательский рейтинг "Actions":', round(array1.mean(),2))
print('средний пользовательский рейтинг "Sports":', round(array2.mean(),2))
print('p-value =', result.pvalue, '\n')

if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")


    
bins = np.linspace(0.1, 10, 21)
plt.figure(figsize=(12,3))
plt.hist(array1, bins, alpha = 0.5, color = 'b')
plt.hist(array2, bins, alpha = 0.5, color = 'g')

plt.plot([array1.mean(), array1.mean()], [0, 150], color = 'b')
plt.plot([array2.mean(), array2.mean()], [0, 150], color = 'lime')

plt.title('Распределение пользовательских оценок')
plt.legend(['mean "action"', 'mean "sports"','dist "action"','dist "sports"'])
plt.xlabel('user score')
plt.ylabel('count')
plt.grid(True)

plt.show()
```
