Компания такси собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.

Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

```python
def decompose (df, start = str(min(df.index)), finish = str(max(df.index))):
    
    temp = df[start : finish]
    decomposed = seasonal_decompose(df[start : finish])

    fig = plt.figure(figsize = (12,5))

    ax1 = fig.add_subplot(411)
    ax2 = fig.add_subplot(412)
    ax3 = fig.add_subplot(413)
    ax4 = fig.add_subplot(414)
    ax1.set_title(f'Seasonal_decompose     {start} - {finish}')
    
    ax1.plot(temp)
    ax2.plot(decomposed.trend, color = 'indianred')
    ax3.plot(decomposed.seasonal, color = 'cadetblue')
    ax4.plot(decomposed.resid, color = 'mediumpurple')
    
    ax1.legend(['data'], loc = 1)
    ax2.legend(['Trend'], loc = 1)
    ax3.legend(['Seasonality'], loc = 1)
    ax4.legend(['Residuals'], loc = 1)
    
    ax1.set_ylim(-30, 320)
    ax2.set_ylim(-30, 320)
    ax3.set_ylim(-125, 125)
    ax4.set_ylim(-125, 125)
    
    for ax in fig.axes:
        ax.grid(True)
        ax.tick_params(axis='x', labelbottom=False, labeltop=False)
        ax.tick_params(axis='y', labelleft=True, labelright=False)
        ax.set_xlim(pd.Timestamp(start), pd.Timestamp(finish))
        ax.set_ylabel('кол-во \nзаказов \nв час')

    ax4.tick_params(axis='x', labelbottom=True, labeltop=False)
    ax4.set_xlabel('дата')

    plt.tight_layout(h_pad = -0.7)

    plt.show()
```

## Анализ

### добавление новых признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    
#    data['year'] = data.index.year
#    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    n = max(max_lag, rolling_mean_size)
    data = data[n:]
    return data
```

### разбивка данных на тренировочную и тестовую выборки

```python
def split_01 (data):
    features = data.drop ('num_orders', axis = 1)
    train = data['num_orders']

    f_train, f_test, t_train, t_test = train_test_split(features, train, shuffle=False, test_size=0.1, random_state = rs)
    return (f_train, f_test, t_train, t_test)
```

### масштабирование данных

```python
def scale (train, test):
    col = list(train)

    scaler = StandardScaler()
    scaler.fit(train[col])

    train[col] = scaler.transform(train[col])
    test[col]  = scaler.transform(test[col])
    
    return(train, test)
```

```python
def cv_time_series_rmse(model, data_f):
    n_splits = 5
    max_train_size = 1000
    tscv = TimeSeriesSplit(n_splits = n_splits, max_train_size = max_train_size)
    rmse = 0.0
    for train_index, test_index in tscv.split(data_f):
        train = data_f.iloc[train_index]
        test = data_f.iloc[test_index]
        f_train = train.drop('num_orders', axis=1)
        t_train = train['num_orders']
        f_test = test.drop('num_orders', axis=1)
        t_test = test['num_orders']
        model.fit(f_train, t_train)
        
        p_test = model.predict(f_test)
        rmse += mean_squared_error(t_test, p_test) ** 0.5
    return rmse / n_splits
```

### модель: линейная регрессия

```python
train, test = train_test_split(df, shuffle=False, test_size=0.1, random_state = rs)
```

```python
%%time
index = range (10, 28)
columns = range (1,40)
rmse_lg = pd.DataFrame(data = [], columns = [], index = index)

model1 = LinearRegression()

for j in tqdm(columns):
    res = []
    for i in index:
        df2 = make_features(train.copy(), i, j)
        rmse = cv_time_series_rmse(model1, df2)
        res.append(rmse)
    rmse_lg[j] = res
```

```python
min_rmse_lg = rmse_lg.values.min()

max_lag_min = min(np.where(rmse_lg == min_rmse_lg)[0]) + min(index)
rolling_mean_size_min = min(np.where(rmse_lg == min_rmse_lg)[1]) + min(columns)

#print('min_rmse =', np.where(rmse_lg == min_rmse_lg))
print(min_rmse_lg)
print('rolling_mean_size =', max_lag_min)
print('max_lag =', rolling_mean_size_min)
```

```python
plt.figure(figsize = (12,3))
plt.plot(rmse_lg, '-')
plt.plot(max_lag_min, min_rmse_lg, 'o', color = 'r')
plt.title('rMSE vs max_lag')
plt.xlabel('max_lag')
plt.ylabel('rmse')
plt.ylim(25,35)
plt.grid(True)
plt.show()
```

```python
res = []
x= []
model1 = LinearRegression()

for i in tqdm(range (1, 30)):
    df2 = make_features(train.copy(), i, rolling_mean_size_min)
    
    rmse = cv_time_series_rmse(model1, df2)
    res.append(rmse)
    x.append(i)
```

```python
plt.figure(figsize = (12,3))
plt.plot(x,res)
plt.title('rMSE vs max_lag')
plt.grid(True)
plt.ylim(25,33)
plt.xlabel('max_lag')
plt.ylabel('rmse')
plt.show()
```

#### Промежуточные выводы:
1. Обучена модель LinearRegression для различных параметров исходных данных:
    - предшествующие значения за последние max_lag часов
    - среднее значение за предшествующие rolling_mean_size часов
2. Значение rMSE от rolling_mean_size имеет периодические минимумы, расположенные на частотах кратных 24-25 часам
3. Значение rMSE от max_lag имеет стапенчатую структуру, с характерным расстоянием в 24-25 часов
4. При значениях max_lag более 30 наблюдаются неусточивости модели. В дальнейшем будеи использовать значение 30 часа.


### Модель: LightGBM

```python
#df2 = make_features(train.copy(), 28, 26)
#df3 = make_features(test.copy(), 28, 26)

df2 = make_features(train.copy(), max_lag_min, rolling_mean_size_min)
df3 = make_features(test.copy(), max_lag_min, rolling_mean_size_min)

f_train = df2.drop('num_orders', axis=1)
t_train = df2['num_orders']

f_test = df3.drop('num_orders', axis=1)
t_test = df3['num_orders']
```

```python
%%time
grid_space = {'n_estimators': [20,50,100],
              'max_depth' : [5,10,15,20],
              'num_leaves': [5,10,20,50],
              'learning_rate': [0.05,0.1,0.2]}


tscv = TimeSeriesSplit(n_splits=5)
model = lgb.LGBMRegressor(n_jobs = -1, random_state = rs)
model_opt = GridSearchCV(estimator=model,
                    param_grid=grid_space,
                    cv=tscv,
                    scoring='neg_root_mean_squared_error')

model_opt.fit(f_train, t_train)
display(model_opt.best_params_)
print('наименьшее rMSE = ', - model_opt.best_score_)
```

##### Промежуточные выводы:
1. Обучена модель LightGBM для различных значениях гиперпараметров
2. Подобрано сочетание гиперпараметров с т.з. минимизации rMSE
3. Достигнутое значение rMSE = 25
