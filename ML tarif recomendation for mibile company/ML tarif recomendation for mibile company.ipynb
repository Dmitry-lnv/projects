## 1 Откройте и изучите файл

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.dummy import DummyClassifier

import random
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/Users/dmitry/Desktop/yandex_python/05. Введение в машинное обучение/users_behavior.csv')

#df = pd.read_csv('/Users/dmitry/Desktop/users_behavior.csv')
```

```python
sns.pairplot(df, hue = 'is_ultra')
plt.gcf().set_size_inches(7,7)

```

## 2 Разбейте данные на выборки

```python
df2, test = train_test_split(df, test_size = 0.2, random_state = 52378342)
train, valid = train_test_split(df2, test_size = 0.25, random_state = 9783421)

train_features = train.drop(['is_ultra'], axis = 1)
train_target = train['is_ultra']
valid_features = valid.drop(['is_ultra'], axis = 1)
valid_target = valid['is_ultra']
test_features = test.drop(['is_ultra'], axis = 1)
test_target = test['is_ultra']
```

## 3 Исследуйте качество разных моделей

### 3.1 Модель DecisionTreeClassifier

```python
def tree_dep_par (param, values):
    x = []
    y = []

    plt.figure(figsize = (10,2))

    for i in values:
        
        if(param == 'max_depth'):
            model = DecisionTreeClassifier(max_depth = i, random_state = 234141)
        elif(param == 'max_features'):
            model = DecisionTreeClassifier(max_features = i, random_state = 1342353)
        elif(param == 'min_samples_leaf'):
            model = DecisionTreeClassifier(min_samples_leaf = i, random_state = 24365)
        elif(param == 'min_samples_split'):
            model = DecisionTreeClassifier(min_samples_split = i, random_state = 65842)
        
        model.fit(train_features, train_target)

        valid_predict = model.predict(valid_features)

        x.append(i)
        y.append(accuracy_score(valid_predict, valid_target))
        #print(i, round(accuracy_score(valid_predict, valid_target),3))

    plt.ylim(0.7, 0.9)
    plt.title(f'accuracy VS {param}')
    plt.grid(True)
    plt.plot(x,y)
    plt.xlabel(f'{param}')
    plt.ylabel('accuracy')
    plt.show()
    
```

```python
tree_dep_par ('max_depth', range(1, 20))
```

```python
tree_dep_par ('max_features', range(1, 5))
```

```python
tree_dep_par ('min_samples_leaf', range(1, 30))
```

```python
# перебираемые значения гиперпараметров установлены на основе 
# рассчитанных зависимостей знаяения точности от варьирования одного из гиперпараметров

grid_space = {'max_depth':[3,5,7,10,15],
              'max_features':[1,2,3],
              'min_samples_leaf':[11,16,20,25]
             }

model =DecisionTreeClassifier()

grid = GridSearchCV(model,
                    param_grid=grid_space,
                    cv=3,
                    scoring='accuracy',
                    verbose=1)
model_grid = grid.fit(train_features, train_target)


print('Параметры:')
display(model_grid.best_params_)
print('Достигнутая точность: ',round((model_grid.best_score_),3))
```

```python
model = model_grid

valid_predict = model.predict(valid_features)
data = confusion_matrix(valid_target, valid_predict)

res_df = pd.DataFrame (data = data,
                      columns = ['predict=0', 'predict=1'],
                      index = ['is_ultra=0','is_ultra=1'])

display(res_df)
print(classification_report(valid_target, valid_predict))
```

### 3.2 Модель RandomForestClassifier

```python
def rf_dep_par (param, values):
    x = []
    y = []

    plt.figure(figsize = (10,2))

    for i in values:
        
        if(param == 'max_depth'):
            model = RandomForestClassifier(max_depth = i, random_state = 43567437)
        elif(param == 'max_features'):
            model = RandomForestClassifier(max_features = i, random_state =1465234)
        elif(param == 'min_samples_leaf'):
            model = RandomForestClassifier(min_samples_leaf = i, random_state = 4764562)
        elif(param == 'min_samples_split'):
            model = RandomForestClassifier(min_samples_split = i, random_state = 14646)
        elif(param == 'n_estimators'):
            model = RandomForestClassifier(n_estimators =i, random_state = 8716454)
        
        model.fit(train_features, train_target)

        valid_predict = model.predict(valid_features)

        x.append(i)
        y.append(accuracy_score(valid_predict, valid_target))
        #print(i, round(accuracy_score(valid_predict, valid_target),3))

    plt.ylim(0.7, 0.9)
    plt.title(f'accuracy VS {param}')
    plt.grid(True)
    plt.plot(x,y)
    plt.xlabel(f'{param}')
    plt.ylabel('accuracy')
    plt.show()
```

```python
rf_dep_par ('max_depth', range(1, 30))
```

```python
rf_dep_par ('max_features', range(1, 5))
```

```python
rf_dep_par ('min_samples_leaf', range(1, 30))
```

```python
rf_dep_par ('n_estimators', range(1, 30))
```

### 3.3 Автоподбор параметров для модели RandomForestClassifier

```python
# перебираемые значения гиперпараметров установлены на основе 
# рассчитанных зависимостей значения точности от варьирования одного из гиперпараметров

grid_space = {'n_estimators':[20, 25],
              'max_depth':[3,5,7,10,15,20],
              'max_features':[1,2],
              'min_samples_leaf':[2,5,10,15,20]
             }

#model = RandomForestClassifier()

grid = GridSearchCV(RandomForestClassifier(),param_grid=grid_space,cv=3,scoring='accuracy',verbose=1)
model_grid = grid.fit(train_features, train_target)


print('Параметры:')
display(model_grid.best_params_)
print('Достигнутая точность: ',round((model_grid.best_score_),2))
```

```python
model = model_grid

valid_predict = model.predict(valid_features)
data = confusion_matrix(valid_target, valid_predict)

res_df = pd.DataFrame (data = data,
                      columns = ['predict=0', 'predict=1'],
                      index = ['is_ultra=0','is_ultra=1'])

display(res_df)
print(classification_report(valid_target, valid_predict))
```
