# Отток клиентов

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры, пороговое значение F1 = 0.59

    
Признаки:
   - RowNumber — индекс строки в данных
   - CustomerId — уникальный идентификатор клиента
   - Surname — фамилия
   - CreditScore — кредитный рейтинг
   - Geography — страна проживания
   - Gender — пол
   - Age — возраст
   - Tenure — сколько лет человек является клиентом банка
   - Balance — баланс на счёте
   - NumOfProducts — количество продуктов банка, используемых клиентом
   - HasCrCard — наличие кредитной карты
   - IsActiveMember — активность клиента
   - EstimatedSalary — предполагаемая зарплата
    
Целевой признак:
   - Exited — факт ухода клиента

## Исследование задачи

```python
def encoding (df, df_mask, columns):
    df2 = df.copy()
    enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
    for col in columns:

        enc.fit(df_mask[[col]])
        for j in list(enc.categories_[0]):
            res = enc.transform(df[[col]])
            col_name = (pd.Series([col]*len(enc.categories_[0])) + 
                        pd.Series(['_']*len(enc.categories_[0])) +
                        pd.Series(enc.categories_[0]))
            res_df = pd.DataFrame (res, columns=col_name)
    
        df2[col_name] = res

    df2 = df2.drop(columns, axis=1)
    return df2
```

```python
num_1 = len(df[df['Exited'] == 1])
num_0 = len(df[df['Exited'] == 0])
print('количество нулевых значений: \t\t', num_0)
print('количество положительнх значений: \t', num_1)
print('их отношений: \t\t\t\t', round((num_0 / num_1), 2))
```

### заполнение NaN в 'Tenure': заполняются одним значением ( i ) или удаление ( i = 11 )

```python
train2, test2 = train_test_split(df, test_size = 0.25, random_state = 78634)

coll_for_encoding = ['Geography', 'Gender']
train = encoding(train2, train2, coll_for_encoding)
test = encoding(test2, train2, coll_for_encoding)

features_train = train.drop(['Exited'], axis=1)
features_test = test.drop(['Exited'], axis=1)

target_train = train['Exited']
target_test = test['Exited']
```

```python
f1_vs_fillna_i = []

num=20

for i in range (0,11,1):
    f1=0
    
    features_train['Tenure'] = features_train['Tenure'].fillna(i)
    features_test['Tenure'] = features_test['Tenure'].fillna(i)
    
    for j in range(0,num,1):
        model = RandomForestClassifier(n_jobs=-1)
        model.fit(features_train, target_train)
        f1 += f1_score(target_test, model.predict(features_test)) / num

    f1_vs_fillna_i.append(f1)

    print(i, f1)
print('f1 = ', max(f1_vs_fillna_i))
```

```python
df2 = df.loc[df['Tenure'].isna() == False]

df2, test3 = train_test_split(df2, test_size = 0.2, random_state = 34598)
train3, valid3 = train_test_split(df2, test_size = 0.25, random_state = 128342)

coll_for_encoding = ['Geography', 'Gender']
train = encoding(train3, train3, coll_for_encoding)
valid = encoding(valid3, train3, coll_for_encoding)
test = encoding(test3, train3, coll_for_encoding)

features_train = train.drop(['Exited'], axis=1)
features_valid = valid.drop(['Exited'], axis=1)
features_test = test.drop(['Exited'], axis=1)

target_train = train['Exited']
target_valid = valid['Exited']
target_test = test['Exited']

print(train.shape, valid.shape, test.shape)
```

## Борьба с дисбалансом

```python
def balance (method, train, random_state):
    train_zeros = train.loc[train['Exited'] == 0]
    train_ones = train.loc[train['Exited'] == 1]
    
    k_0= len(train_zeros)
    k_1= len(train_ones)
    k= k_0/k_1
    
    if k > 1:
        if (method == 'increase'):
            train_upsampled = pd.concat([train_zeros] + [train_ones] * int(round(k,0)))
            print('увеличено кол-во "1"')
        elif (method == 'decrease'):
            train_upsampled = pd.concat([train_zeros.sample(frac= 1/k, random_state=random_state)] + [train_ones])
            print('уменьшего кол-во "0"')
    
    elif k < 1:
        if (method == 'increase'):
            train_upsampled = pd.concat([train_zeros] * int(round(k,0)) + [train_ones])
            print('увеличено кол-во "0"')
        elif (method == 'decrease'):
            train_upsampled = pd.concat([train_ones.sample(frac= 1/k, random_state=random_state)] + [train_zeros])
            print('уменьшего кол-во "1"')
            
    elif k == 1:
        train_upsampled = train
        print('баленсировка не требуется')
        

    train_upsampled = shuffle(train_upsampled, random_state=random_state)

    features_upsampled = train_upsampled.drop(['Exited'], axis=1)
    target_upsampled = train_upsampled['Exited']

    print('кол-во "0" = ', len(train_upsampled.loc[train_upsampled['Exited'] == 0]))
    print('кол-во "1" = ', len(train_upsampled.loc[train_upsampled['Exited'] == 1]))
    return features_upsampled, target_upsampled
```

### Балансировка выборки через увеличение кол-ва событий с результатом 1


```python
f1_best=0
model_best = RandomForestClassifier(random_state=932457, n_jobs=-1)
balance_type = ''
```

```python
features_upsampled_inc, target_upsampled_inc = balance('increase', train, 243622)
```

```python
f1_valid = []

for i in range(1,20,1):
    model_inc = RandomForestClassifier(max_depth = i, n_estimators=120, random_state=932457, n_jobs=-1)
    model_inc.fit(features_upsampled_inc, target_upsampled_inc)
    
    f1 = f1_score(target_valid, model_inc.predict(features_valid))
    
    if f1 > f1_best:
        f1_best = f1
        model_best = model_inc
        balance_type = 'increase'
                  
    f1_valid.append(f1)
 
print('f1_inc = ', max(f1_valid))
print('\n f1_best = ', f1_best)
print('model_best = ', model_best)
print('balance_type = ', balance_type)
```

```python
plt.figure(figsize=(12,3))
plt.plot(range(1,20,1), f1_valid)

plt.grid(True)
plt.title('f1 vs max_depth в случае балансировки выборки через увеличение кол-ва "1"')
plt.show()
```

### Балансировка выборки через сокращение кл-ва событий с результатом 0

```python
features_upsampled_dec, target_upsampled_dec = balance('decrease', train, 902745)
```

```python
f1_valid = []

for i in range(1,20,1):
    model_dec = RandomForestClassifier(max_depth = i, n_estimators=120, random_state=894267, n_jobs=-1)
    model_dec.fit(features_upsampled_dec, target_upsampled_dec)
    
    f1 = f1_score(target_valid, model_dec.predict(features_valid))
    
    if f1 > f1_best:
        f1_best = f1
        model_best = model_dec
        balance_type = 'decrease'
                  
    f1_valid.append(f1)
 
print('f1_dec = ', max(f1_valid))
print('\n f1_best = ', f1_best)
print('model_best = ', model_best)
print('balance_type = ', balance_type)
```

```python
plt.figure(figsize=(12,3))
plt.plot(range(1,20,1), f1_valid)

plt.grid(True)
plt.title('f1 vs max_depth в случае балансировки выборки через уменьшения кол-ва "0"')
plt.show()
```

### Промежуточные выводы:
лучшие результыта метрики f1 обеспечивается балансировкой тренировочных данных с увеличением кол-ва "1"

достигнутое значение f1 = 0.609
